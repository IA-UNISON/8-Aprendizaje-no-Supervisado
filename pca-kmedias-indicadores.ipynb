```python
import numpy as np
import pandas as pd

url_csv = 'indicadores_bienestar_2014.csv'

df = pd.read_csv(url_csv, index_col=0, thousands=',')
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df)
```

```python
estados_pca = pca.transform(df)
estados_pca
```

```python
estados_pca_df = pd.DataFrame(estados_pca)
estados_pca_df.index = df.index
estados_pca_df.columns = ["CP1","CP2"]
estados_pca_df.head()
```

```python
print(pca.explained_variance_ratio_) 
```

El primero componente explica el 95.2% de la varianza, mientras que el segundo componente explica el 4.7% de la varianza. Sumados entre los dos explican el 99.7% de los datos lo cual es muy buen resultado.

```python
%matplotlib inline

ax = estados_pca_df.plot(kind='scatter', x='CP1', y='CP2', figsize=(16,8))

for i, pais in enumerate(estados_pca_df.index):
    ax.annotate(pais, (estados_pca_df.iloc[i].CP1, estados_pca_df.iloc[i].CP2))
```

### K - Medias

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

maxK = 6
listaK = range(2,maxK + 1)
costos=[]

for k in listaK:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    costos.append(-kmeans.score(df))
    
plt.figure(figsize=(16,8))
plt.plot(listaK, costos, '-or')
plt.xlabel('Número de clusters')
plt.ylabel('Costo')
plt.title('Selección de K con el método del codo')
```

El codo mas pronunciado se encuentra con 3 clusters
